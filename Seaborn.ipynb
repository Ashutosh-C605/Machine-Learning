```python
import seaborn as sns
# %matplotlib inline #to  make the plot appear just below the code cell not required in this version by default it is true

tips = sns.load_dataset('tips')
print(tips.head(50))
# sns.distplot([0,1,2,3,4,5], hist=False)
sns.histplot([0,1,2,3,4,5], kde=True)
```

```python
sns.jointplot(x="total_bill", y="tip",data=tips, kind="hex",color='red')
```

```python
tips[tips['size']==4].value_counts()
```

```python
sns.pairplot(tips)    #in total bill vs total bill height of the histogram shows the frequency
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

```python
sns.barplot(x='sex', y='total_bill',data=tips, estimator='max')
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue="smoker")
```

```python
sns.violinplot(x='day',y='total_bill',hue='smoker' ,data=tips)
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='strip')
```
